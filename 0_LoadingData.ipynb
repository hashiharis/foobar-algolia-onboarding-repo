```python
#Formatting the data to Algolia acceptable JSON Format
csv_file = pd.DataFrame(pd.read_csv("Ikea_data.csv", sep = ",", header = 0, index_col = False))
csv_file=csv_file.drop(['Unnamed: 0'],axis=1)
csv_file.to_json("Ikea_data.json", orient = "records", date_format = "epoch", double_precision = 10, force_ascii = True, date_unit = "ms", default_handler = None)
```

```python
# Upload data to Algolia index
from algoliasearch.search_client import SearchClient

#Start the search client

client = SearchClient.create('testingRR289Y7D9N','7d86858c421afc48c26562605348cdb8')
```

```python
#Create an index
import json

index = client.init_index('Ikea_testdata')
batch=json.load(open('Ikea_data.json'))
# print(batch)
index.save_objects(batch,{'autoGenerateObjectIDIfNotExist': True})
```
