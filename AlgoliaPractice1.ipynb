```python
# Opening the CSV data and previewing

import pandas as pd

data=pd.read_csv('Ikea_data.csv')
data
```

```python
#Formatting the data to Algolia acceptable JSON Format
csv_file = pd.DataFrame(pd.read_csv("Ikea_data.csv", sep = ",", header = 0, index_col = False))
csv_file=csv_file.drop(['Unnamed: 0'],axis=1)
csv_file.to_json("Ikea_data.json", orient = "records", date_format = "epoch", double_precision = 10, force_ascii = True, date_unit = "ms", default_handler = None)
```

```python
# Upload data to Algolia index
from algoliasearch.search_client import SearchClient

#Start the search client

client = SearchClient.create('testingRR289Y7D9N','7d86858c421afc48c26562605348cdb8')


```

```python
#Create an index
import json

index = client.init_index('Ikea_testdata')
batch=json.load(open('Ikea_data.json'))
# print(batch)
index.save_objects(batch,{'autoGenerateObjectIDIfNotExist': True})
```

```python
# Setting the searchable attributes
index.set_settings({
    'searchableAttributes':[
        'unordered(name)',
        'unordered(designer)',
        'unordered(category)',
        
    ]
})
```

```python
#Setting the custom ranking attribute
index.set_settings({
    'customRanking':[
        'asc(price)'
    ]
})
```

```python
#Searching for the first 20 search results for RÖDEBY

res = index.search('chair', {
    'attributesToRetrieve': [
        '*'
    ],
    'hitsPerPage': 20
}) 

```

```python
res=pd.DataFrame(data=res)
res=pd.DataFrame(data=res['hits'])
res['hits']
```

```python
load_items = []

for item in res['hits']:
    load_items.append({
        'name':item['name'],
        'category':item['category'],
        'price':item['price'],
        'objectID':item['objectID'],
        'old_price':item['old_price'],
        'sellable_online':item['sellable_online'],
        'link':item['link'],
        'other_colors':item['other_colors'],
        'short_description':item['short_description'],
        'designer':item['designer'],
        'depth':item['depth'],
        'height':item['height'],
        'width':item['width']
    })
res=pd.DataFrame(load_items)
res
```
